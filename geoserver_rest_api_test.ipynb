```python
# CODE TO CREATE A FEATURE STORE
# GIVE THE SAME NAME TO THE PG_TABLE PARAMETER AS THAT OF THE SHAPEFILE NAME -- VERY IMPORTANT

"""
name = datastore name
path = path of the shapefile you want to publish
workspace = name of the workspace created earlier
pg_table = name of the shapefile to be uploaded (this should be same as the name of the shapefile in the path mentioned)
"""

geo.create_datastore(name="tourist_locations", path=r'/Users/saimanojappalla/Downloads/audiotour/survey/survey.shp', workspace='geoserver-rest-test')
geo.publish_featurestore(workspace='geoserver-rest-test', store_name='tourist_locations', pg_table='survey')

```
